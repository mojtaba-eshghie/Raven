```python
import matplotlib.pyplot as plt

# Data
labels = [
    "revert",
    "require",
    "no source code found",
    "out of gas",
    "failures",
    "assert",
    "arith. under/overflow",
    "div/mod by zero",
]
values = [6033, 5660, 4660, 2420, 600, 529, 84, 14]

# Style for compact IEEE figure
plt.rcParams.update({
    "font.size": 7,
    "figure.figsize": (3.4, 1.6),
    "axes.spines.top": False,
    "axes.spines.right": False,
    "savefig.pad_inches": 0,
})

fig, ax = plt.subplots()
bars = ax.barh(labels, values, color="#4B9CD3", height=0.5)
ax.invert_yaxis()
ax.grid(axis="x", linestyle=":", linewidth=0.4, alpha=0.7)

ax.set_xlabel("Count", fontsize=8)
ax.set_ylabel("")
ax.tick_params(axis='y', pad=1, length=0)
ax.xaxis.set_tick_params(pad=1)

# Add value labels
for i, v in enumerate(values):
    ax.text(
        v + 50,  # offset slightly to the right of the bar
        i,       # same y as bar index
        f"{v:,}",  # formatted with commas
        va="center",
        ha="left",
        fontsize=7,
    )

# Remove extra space
plt.subplots_adjust(left=0.28, right=0.98, top=0.98, bottom=0.15)
plt.savefig("failure_invariant_summary.pdf", bbox_inches="tight", pad_inches=0)
plt.close()
```
